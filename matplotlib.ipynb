#### Data Visualization with Matplotlib (More Basic)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('dark_background')
```

```python
roll_no = [2, 4, 5, 6, 7, 8]
marks = [10,20,30,40,50,60]
```

```python
plt.scatter(roll_no,marks,color='gray',marker='.')
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.scatter(roll_no,marks,color='blue',marker='.')
plt.show()
```

```python
plt.plot(roll_no,marks,'b*',markersize=15)
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.xticks(np.arange(0,60,5))
plt.yticks(np.arange(10,60,5))
plt.plot(roll_no,marks,'bo')
plt.xlabel('Roll numbers')
plt.ylabel('marks of student')
plt.show()
```

### Line PLot
Connecting the dots of scatter

```python
plt.plot(roll_no,marks,linestyle='-.')
```

```python
plt.plot(roll_no,marks,linestyle=':',linewidth=5)
```

```python
colors=['red','blue','indigo','cyan','orange']
plt.figure(figsize=(10,6))
plt.bar(roll_no,marks,color = colors,width=0.6)
```

## Pie Chart


```python
subjects = ["Math", "Science", "English", "History"]
marks = [85, 90, 78, 88]
values=[0.1,0.2,0,0]
plt.pie(marks,labels=subjects,autopct='%0.2f%%',explode=values)
plt.show()
```
