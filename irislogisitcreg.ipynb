### Iris dataset model using Logistic Regression

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
thedata = load_iris()
```

```python
samples = [10,25,80]
```

```python
thedata.target[samples]
```

```python
thedata.target_names[thedata.target[samples]]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train,x_test,y_train,y_test = train_test_split(thedata.data,thedata.target,random_state=2,test_size=0.2)
```

```python
x_train.shape, y_train.shape
```

```python
lor = LogisticRegression()
```

```python
lor.fit(x_train,y_train)
```

```python
ypred = lor.predict(x_test)
```

```python
lor.score(x_test,y_test)
```

```python
import seaborn as sb
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,ypred)
cm
```

```python
plt.Figure(figsize=(10,7))
```

```python
sb.heatmap(cm,annot=True)
```
