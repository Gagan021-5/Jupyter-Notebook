```python
import pandas as  pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv('ageincome.csv')
df.head()
```

```python
plt.scatter(df['Age'],df['Income'])
```

```python
km = KMeans(n_clusters=3)
km
```

```python
ypred = km.fit_predict(df[['Age','Income']])
```

```python
df['cluster'] = ypred
df.head()
```

```python
df0 = df[df.cluster == 0]
df1 = df[df.cluster == 1]
df2 = df[df.cluster == 2]
plt.scatter(df0.Age,df0['Income'],color='green')
plt.scatter(df1.Age,df1['Income'],color='red')
plt.scatter(df2.Age,df2['Income'],color='black')

plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()
```

```python
scaler = MinMaxScaler()
```

```python
km = KMeans(n_clusters=3)
ypredd = km.fit_predict(df[['Age','Income']])
ypredd
```

```python
df['Cluser'] = ypredd
```

```python
df.drop('cluster',axis='columns',inplace=True)
```

```python
df
```

```python
df0 = df[df.Cluser == 0]
df1 = df[df.Cluser == 1]
df2 = df[df.Cluser == 2]
plt.scatter(df0.Age,df0['Income'],color='green')
plt.scatter(df1.Age,df1['Income'],color='red')
plt.scatter(df2.Age,df2['Income'],color='black')

plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()
```
