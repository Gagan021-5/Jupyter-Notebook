```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
taxidf = sb.load_dataset('taxis')
taxidf.head()
```

#### Scatter Plotting fxn

```python
sb.scatterplot(x='fare',y='distance',data=taxidf,hue='tip')
plt.title("Fare vs Distance for Taxis data")
```

```python
sb.scatterplot(x='fare',y='distance',data=taxidf,hue='tip',style='dropoff_zone',legend=False,palette='inferno',alpha=0.3)
sb.lineplot(x='fare',y='distance',data=taxidf,palette='brown')
```

## BarPlots (Bargraph)

```python
sb.barplot(x='fare',y='distance',data=taxidf.head(),hue='passengers')
```

```python
#orient
sb.barplot(x='fare',y='distance',data=taxidf.head(),hue='passengers',
           orient='y')
```
