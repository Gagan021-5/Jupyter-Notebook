```python
# %%
import pandas as pd

# %%
from sklearn.datasets import load_iris
iris  = load_iris()

# %%
dir(iris)

# %%
iris.feature_names

# %%
df = pd.DataFrame(data= iris.data,columns= iris.feature_names)

# %%
df.head()

# %%
df['target'] = iris.target

# %%
iris.target_names

# %%
df.head()

# %%
df[df.target == 2].head()

# %%
df['flower'] = df.target.apply(lambda x: iris.target_names[x])

# %%
df.head()

# %%
import matplotlib.pyplot as plt

# %%
%matplotlib inline

# %%
df0 = df[df.target==0]
df1 = df[df.target ==1]
df2 = df[df.target ==2]

# %%
df0.head()

# %%
df1.head()

# %%
df2.head()

# %%
plt.xlabel('Sepal Length')
plt.ylabel('Sepal width')
plt.scatter(x=df0['sepal length (cm)'],y=df1['sepal width (cm)'],color='blue',marker='+')
plt.scatter(x=df1['sepal length (cm)'],y=df2['sepal width (cm)'],color='green',marker='.')

# %%
plt.scatter(x=df0['petal length (cm)'],y=df1['petal width (cm)'],color='blue',marker='+')
plt.scatter(x=df1['petal length (cm)'],y=df2['petal width (cm)'],color='green',marker='.')

# %%
from sklearn.model_selection import train_test_split

# %%
x = df.drop(['target','flower'],axis='columns')

# %%
x.head()

# %%
y = df.target

# %%
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=1,test_size=0.2)

# %%
x_train.shape,y_train.shape




```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```
