```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
input = df[['Pclass','Sex','Age','Fare']]
target = df['Survived']
```

```python
input.head()
```

```python
target.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_sex = LabelEncoder()
input['sex_n'] = le_sex.fit_transform(input['Sex'])
```

```python
input.head()
```

```python
input_n = input.drop(['Sex'],axis='columns')
input_n.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(input_n,target,random_state=42,test_size=0.2)
```

```python
x_train.shape,y_train.shape
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
```

```python
model = dt.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[3,22.0,7.25,1]])
```
