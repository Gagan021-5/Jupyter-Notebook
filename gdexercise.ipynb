```python
import numpy as np
```

```python
def gd(x,y):
    m = b = 0
    epochs = 10
    n = len(x)
    lr = 0.001
    for i in range(epochs):
        y_pred = m * x + b
        costfn = (1/n) * sum([val **2 for val in (y-y_pred)])
        md = -(2/n) * sum(x*(y-y_pred))
        bd = -(2/n) * sum(y-y_pred)
        m = m - lr * md
        b = b - lr * bd
        print("m {},b {},,cost {}, iteration {}".format(m,b,costfn,epochs)) 
```

```python
x = np.array([92,56,88,70,80,49,65,35,66,67])
y = np.array([98,68,81,80,83,52,66,30,68,73])
```

```python
gd(x,y)
```
