```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(data=iris.data,columns=iris.feature_names)
```

```python
df.head()
```

```python
newdf = df.drop(df[['sepal length (cm)','sepal width (cm)']],axis='columns',inplace=True)
newdf
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=2)
```

```python
ypred = km.fit_predict(df[['petal length (cm)','petal width (cm)']])
```

```python
df['cluster'] = ypred
```

```python
df.sample(6)
```

```python
df0 = df[df.cluster == 0]
df1 = df[df.cluster == 1]
```

```python
plt.scatter(df0['petal width (cm)'],df0['petal length (cm)'],color='red',marker='.')
plt.scatter(df1['petal width (cm)'],df1['petal length (cm)'],color='indigo',marker='.')
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()

ypred = df[['petal length (cm)', 'petal width (cm)']] 
ypred = df[['petal length (cm)', 'petal width (cm)']] = scaler.fit_transform(
    df[['petal length (cm)', 'petal width (cm)']]
)
```

```python
df['cluster'] = ypred
```

```python
df.head()
```

```python
df0 = df[df.cluster == 0]
df1 = df[df.cluster == 1]
plt.scatter(df0['petal width (cm)'],df0['petal length (cm)'],color='red')
plt.scatter(df1['petal width (cm)'],df1['petal length (cm)'],color='indigo')
plt.xlabel("Petal width")
plt.ylabel("Petal length")
plt.legend()
```
